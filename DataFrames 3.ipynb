# Filtering with the WHERE Method:

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

### So the WHERE method is unique because it's actually going to return the original data frame, the complete data frame, even though we're filtering. 
### Those rows that have a matching condition or those roles that are valid for our condition are going to have all of their values represented and present. 
### The rows that do not match our criteria are just going to be filled with null values.

```python
# Lets create a booleon series, ie create a critera:
# Extract value in the actor column that is equal to Sean Connery.
```

```python
mask = bond["Actor"] == "Sean Connery"
bond[mask].head(3)
```

```python
bond.where(mask).head()
```

```python
bond.where(bond["Box Office"] > 800).head()
```

```python
mask2 = bond["Box Office"] > 800
```

```python
bond[mask & mask2]
```

```python
bond.where(mask&mask2).head(3)
```

# .apply() Method on single columns

```python
import pandas as pd
```

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

### Apply is usually used for something like a more custom operation, especially if you're doing things like conditionals. If this, give me that. Otherwise, give me that.

```python
def convert_to_string_and_add_millions(number):
    return str(number) + " MILLIONS!"

bond["Box Office"] = bond["Box Office"].apply(convert_to_string_and_add_millions)
```

```python
bond["Budget"] = bond["Budget"].apply(convert_to_string_and_add_millions)
```

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + " MILLIONS!"
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
```

```python
bond.head(3)
```

# The .apply() Method with Row Values:

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
    
bond.apply(good_movie, axis = "columns")
```

#### This method changes the actual df (bond). to avoid this we cna create a copy with all changes and use that:

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond.head(3)
```

```python
directors = bond["Director"].copy()
directors.head(3)
```

```python
directors["A View to a Kill"] = "Mister John Glen"
```

```python
directors.head(3)
```

```python
bond.head(3)
```

### Copy Method is helpful:
### All it does is create the exact same thing. It just stores it in a seperate place in memory.
### Modification to copied object does not affect the original object from which they are copied.
