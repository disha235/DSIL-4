# Create a MultiIndex with set_index Method:

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates=["Date"])
bigmac.head(2)
```

```python
# Using the keys="string" that string will become the index now:

bigmac.set_index("Date").head(2)
```

```python
# It is keys (plural) so we can use multiple strings: 
bigmac.set_index(keys = ["Date", "Country"]).head(3)
```

```python
bigmac.set_index(keys = ["Country", "Date"]).head(3)
```

### Which column to put on the outermost layer? 
#### General Best Practice is to take the column with the least number of unique values and put it in the most outer level and continue taht relationship as you proceed downwards.

```python
# here we have less unique dates than country:
bigmac.nunique()
bigmac.set_index(keys = ["Date", "Country"], inplace=True)
```

```python
bigmac.sort_index()
bigmac.sort_index(inplace = True)
bigmac.head(2)
```

# Extract Index Level Values with the get_level_values Method:

### Get values from any given values of index:

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.head(3)
```

```python
bigmac.index.get_level_values("Date")
bigmac.index.get_level_values(0)
```

```python
bigmac.index.get_level_values("Country")
bigmac.index.get_level_values(1)
```

# Change Index Level Name with the set_names Method:

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
bigmac.index.set_names(names = ["Day", "Location"], inplace = True)
bigmac.head(3)
```

```python
bigmac.index.set_names(names = "Date", level = 0)

bigmac.index.set_names(names = "Date", level = "Day", inplace = True)
```

# The TRANSPOSE Method:

```python
bigmac =pd.read_csv("bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head()
```

```python
bigmac = bigmac.transpose()
```

```python
bigmac.loc[("Price in US Dollars",), ("2010-01-01","Sri Lanka")]
```

```python
bigmac.loc[("Price in US Dollars",), ("2010-01-01","Sri Lanka"):("2010-01-01","Ukraine")]
```

# The swaplevel Method:

```python
bigmac =pd.read_csv("bigmac.csv", parse_dates=["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
bigmac.swaplevel()
```

```python
bigmac.swaplevel(0, 1)
bigmac.swaplevel(1, 0)

bigmac.swaplevel("Date", "Country")
bigmac.swaplevel("Country", "Date")
```

```python
bigmac = bigmac.swaplevel(1, 0)
bigmac.head(3)
```

# The Pivot Method:
### Used to reorient a data set. Take values in the column and convert into column head.

```python
import pandas as pd
```

```python
pd.read_csv("salesmen.csv")
output=None
```

```python
sales = pd.read_csv("salesmen.csv", parse_dates=["Date"])
len(sales)
```

```python
sales["Salesman"].value_counts()
```

```python
# All are 366 values, so we can change the Salesman into category.
```

```python
sales = pd.read_csv("salesmen.csv", parse_dates=["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
```

```python
# Has lot of repeating format. Like 366 times Rob, Ronald, etc keeps repeating.
# So we can pivot to a format that is more readable and saves space.
```

```python
# 5 unique values of Salesman chaneg to column headers
# convert revenue values as values for each one of those names:
```

```python
# sales.pivot(index, columns, values)
sales.pivot(index = "Date", columns = "Salesman", values = "Revenue")
```

#  The pivot_table() Method:
### It allows to aggregate the data frame/data sets to look at the values as a whole.

```python
foods = pd.read_csv("foods.csv")
```

```python
# get info for Sales based on Gender:

foods.pivot_table(values = "Spend", index = "Gender", aggfunc = "mean")
```

```python
foods.pivot_table(values = "Spend", index = "Gender", aggfunc = "sum")
```

```python
# Get info for the Spend based on Item sale:
foods.pivot_table(values = "Spend", index = "Item", aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = "Gender", aggfunc = "mean")
```

```python
# Can provide List to that index parameter
# This will help create a multiindex parameter:
```

```python
# Group Gender and then by Items:

foods.pivot_table(values = "Spend", index = ["Gender", "Item"], aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = ["Frequency", "City"], aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "count").head(3)
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "max").head(3)
```

```python
pd.pivot_table(data = foods, values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "sum")
```

#  The pd.melt() Method:

```python
# Create one line for Boris, q1/2/3 and 4:
```
