# So what exactly is a data frame?

A series is a one dimensional data structure.

Fundamentally, a series is just a single column of data.

In comparison, a data frame is a two dimensional data structure, two dimensional meaning, it consists of rows and columns.

So the easiest way to think about a data frame is as a regular table, the kind you might see in a spreadsheet, or like the circle table of a database. It's a combination of rows and columns.


When we say that the data stream is two dimensional, we don't really care about the number of rows and columns.
We care about the number of points of reference that we need in order to extract any single value from that data structure.

# Why is the series a one dimensional data structure?

Because we only need one identifier or one reference point to extract any given value.

For example, if I have a series and I say give me the value at row 435, that reference of 435 is enough information for me to get that value.

So I need only one piece of information, the row position or the row, whatever, to get a value.

So it's a one dimensional data structure.

Now compare that to a table that has multiple rows and multiple columns.

If I were to say, give me the value row 435, that is not enough information to extract a specific value because you don't know which column you are referring to.

So in a table with rows and columns, we need two points of reference. We need the row and we need the column. We need both so that we can find the intersection where the two meet and that cell is our given data point.

So because we need two points of reference, it is a two dimensional dataset.

The number of rows and columns don't matter if we had a million rows and a million columns, we'd still need two points of reference, a row identifier and a column identifier. So the number of rows and columns doesn't matter.

What matters is how many points of reference we need to get a value, and that logic continues infinitely.

# So what is a three dimensional dataset?

It's one where you need three points of reference in order to extract a value.

So imagine we had two tables and both of those tables had rows and columns in order to extract a given value.

I need to give you first the table and then the row and the column within that table to get a value.

So that's three points of reference so that entire container of data can be called three dimensional.

# Sort a DataFrame with the .sort_values() Method:

```python
nba = pd.read_csv("nba.csv")
nba.tail(3)
```

```python
# nba.sort_values()
# Parameter for sort_values:
# by: expecting a string of the column that we want to sort by
```

```python
# Will sort the Names column in Alphabetical order.
nba.sort_values("Name")
```

```python
# Will sort the Names column in descending order (Z to A).
nba.sort_values("Name", ascending = False)
```

```python
# Will sort by the Age Column. In a Descending order:
nba.sort_values("Age", ascending=False)
```

```python
# Sort by Salaries column. Can look at the highest salasry:
nba.sort_values("Salary", ascending = False)
```

### When sort.values method is used null values are at the very end of the column.

```python
nba.sort_values("Salary").tail()
```

### Null Values at the end is due to: na_position = "last" 

```python
# if we want null values to be first, 
# while still keeping ascending or descending order

nba.sort_values("Salary", na_position="first")
```

# Sort a DataFrame with the .sort_values() Method for Multiple Columns:

```python
nba = pd.read_csv("nba.csv")
nba.tail(3)
```

```python
# First sorts with the Team column alphabetically
# then sorts by the Name column alphabatically:

nba.sort_values(["Team", "Name"])
```

```python
# Will execute both in reverse-alphabetical order
# First the teams column is sorted and then the name column:

nba.sort_values(["Team", "Name"], ascending = False)
```

```python
# Will execute both in reverse-alphabetical order
# First the teams column is sorted and then the name column:

nba.sort_values(["Team", "Name"], ascending = [True, False])

# will assign boolean sequentially to taht sort:
# ie True for Team and False for Name

nba.sort_values(["Team", "Name"], ascending = [True, False], inplace = True)
nba.head(3)
```

# Rank Values with the .rank() Method:

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
# For rank method to work we need to drop all the null values:
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head(3)
```

### Rank Salaries in NBA:

```python
nba["Salary"].rank(ascending = False)
```

```python
# Ranked salaries as integers:

nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
nba.head(3)
```

```python
nba.sort_values(by = "Salary", ascending = False)
```
